## Análisis de los Datos con Pandas
Comenzamos con el dataset de Escuelas en la República Argentina. Analizamos la información de las columnas.

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/LucasSFlores80/Data-Science-I/main/escuelas.csv'
data_3 = pd.read_csv(url, encoding='latin-1')
data_3
```

```python
# Vemos los datos de las columnas
data_3.columns
```

Analizamos las escuelas de la provincia de Mendoza

```python
data_3_mza = data_3[data_3['Jurisdicción']=='Mendoza']
data_3_mza
```

Nos fijamos cuales son estatales, rurales y de educación técnica

```python
data_3_mza_est_rural = data_3_mza[(data_3_mza['Sector']=='Estatal') & (data_3_mza['Ámbito']=='Rural') & (data_3_mza['Secundaria Técnica (INET)']=='X')]
data_3_mza_est_rural
```
